```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## Simulated I–V Data

In many materials experiments, current–voltage (I–V) measurements are used to study
electrical transport properties. Here, we simulate a simple Ohmic response with small
experimental noise.

```python
# Voltage range (Volts)
V = np.linspace(-5, 5, 50)

# Ohmic current with noise (I = aV + noise)
a = 0.002  # conductance (A/V)
noise = np.random.normal(0, 0.0002, size=len(V))
I = a * V + noise
```

## Plotting the I–V Characteristics

```python
plt.figure(figsize=(6,4))
plt.plot(V, I, 'o', label='Experimental data')
plt.xlabel("Voltage (V)")
plt.ylabel("Current (A)")
plt.title("Simulated I–V Characteristics")
plt.legend()
plt.grid(True)
plt.show()
```

## Linear Fit and Conductance Estimation

For an Ohmic material, current varies linearly with voltage. The slope of the I–V curve
gives the electrical conductance.

```python
# Linear fit
slope, intercept, r_value, p_value, std_err = stats.linregress(V, I)

# Fitted current
I_fit = slope * V + intercept
```

```python
plt.figure(figsize=(6,4))
plt.plot(V, I, 'o', label='Experimental data')
plt.plot(V, I_fit, '-', label='Linear fit')
plt.xlabel("Voltage (V)")
plt.ylabel("Current (A)")
plt.title("I–V Curve with Linear Fit")
plt.legend()
plt.grid(True)
plt.show()
```

## Extracted Parameters

```python
print(f"Conductance (slope): {slope:.4e} A/V")
print(f"Intercept: {intercept:.4e} A")
print(f"Correlation coefficient (R): {r_value:.4f}")
```

## Physical Interpretation

The linear I–V behavior indicates Ohmic conduction, which is expected for a uniform
conducting material with good electrical contacts. The extracted conductance is consistent
with the simulated value, and the small deviation arises from experimental noise.

Such analysis is routinely used in thin-film and nanomaterials research to evaluate
electrical transport properties.
